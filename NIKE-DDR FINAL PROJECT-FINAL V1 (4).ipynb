```python
#################### OBTANING VARIOUS ATTRIBUTES FOR ALL SALE SHOES #####################

import os
import re
from bs4 import BeautifulSoup

path = 'nike_links3'  # saved html files path

# List to hold dictionaries of product info
nike_products_list = []  

# Looping through each file in the directory
for file_name in os.listdir(path):
    if file_name.endswith(".html"):
        filepath = os.path.join(path, file_name)
        with open(filepath, 'r', encoding='utf-8') as file:
            html_content = file.read()

        soup = BeautifulSoup(html_content, 'html.parser')

        # Dictionary to store product info
        product_info = {}

        #  product title
        title_element = soup.select_one('h1')
        title = title_element.text.strip() if title_element else "NULL"
        print("Product title is:", title)
        product_info['title'] = title

        # print category
        category_element = soup.select_one('h2')
        category = category_element.text.strip() if category_element else "NULL"
        print("Footwear Style Category is:", category)
        product_info['category'] = category

        #  sale price
        sale_price_element = soup.find(attrs={"data-test": "product-price-reduced"})
        sale_price = sale_price_element.get_text().strip() if sale_price_element else "NULL"
        print("Product Sale Price is:", sale_price)
        product_info['sale_price'] = sale_price

        # original price
        original_price_element = soup.find(attrs={"data-test": "product-price"})
        original_price = original_price_element.get_text().strip() if original_price_element else "NULL"
        org_price_match = re.search(r'\$\d+', original_price)
        org_price = org_price_match.group() if org_price_match else "NULL"
        print("Original Product price is:", org_price)
        product_info['original_price'] = org_price

        #  discount offer
        offer_element = soup.find(attrs={"data-testid": "OfferPercentage"})
        if offer_element:
            offer_perc = offer_element.get_text().strip()
            offer_perc_final = offer_perc.split()[-2] if offer_perc else "NULL"
            print("Discount Offer on the product is:", offer_perc_final)
        else:
            print("No discount offer available for this product.")
        product_info['discount_offer'] = offer_perc_final if offer_element else "NULL"

        # description
        desc_element = soup.select_one('.description-preview.body-2.css-1pbvugb > p')
        desc_final = desc_element.get_text(strip=True) if desc_element else "NULL"
        print("Description is:", desc_final)
        product_info['description'] = desc_final

        # product ID
        element = soup.select_one('.description-preview__style-color.ncss-li')
        style_id = element.text.strip()[-11:] if element else "NULL"
        print("The Product ID for the product is:", style_id)
        product_info['product_id'] = style_id

        #  number of reviews
        review_span = soup.select_one('h3.headline-4 > span')
        number_of_reviews = review_span.get_text(strip=True).split('(')[-1].strip(')') if review_span else "NULL"
        print("Number of reviews is:", number_of_reviews)
        product_info['number_of_reviews'] = number_of_reviews if review_span else "0"

        # rating
        rating_div = soup.find('div', class_='css-n209rx')
        rating = rating_div.get('aria-label') if rating_div else "NULL"
        print("Rating is:", rating)
        product_info['rating'] = rating

        # Add the product_info dictionary to the list
        nike_products_list.append(product_info)

        print("*****************************************************************************************************")
```

## Data Wrangling

Manipulating the data to make it usable for analysis

If a title contains a collection name, that name will be inserted into the 'collection' column. If not, 'others' will be used to signify that it doesn't match any of the known collections.

```python
import mysql.connector

SQL_DB = "nike"   
SQL_TABLE_NIKE = "products"

try:
    #connect to server
    conn = mysql.connector.connect(host='localhost',
                                   user='root',
                                   password='Rach@76732',
                                   )
    cursor = conn.cursor()
    cursor.execute('USE nike')

    #  if 'collection' column already exists then nochange and add it if it doesn't alter table
    cursor.execute("SELECT * FROM INFORMATION_SCHEMA.COLUMNS WHERE TABLE_SCHEMA = %s AND TABLE_NAME = 'products' AND COLUMN_NAME = 'collection'", (SQL_DB,))
    result = cursor.fetchone()
    if not result:
        query = "ALTER TABLE " + SQL_TABLE_NIKE + " ADD COLUMN collection VARCHAR(255)"
        cursor.execute(query);
        
     # Select all product titles from the table
    cursor.execute(f"SELECT product_id, title FROM products;")
    products = cursor.fetchall()
        

    # Checking for Collection name in the title
        
    # first update collection with 'others'
    for product_id, title in products:
        collection_name = 'others'
        title = title.strip()  
    
    # Check if any of the known collection names are in the title, if available then update collection column
        for collection in collection_names:
            if collection.lower() in title.lower():
                collection_name = collection
                break
            
            
        update_query = f"UPDATE products SET collection = %s WHERE product_id = %s;"
        cursor.execute(update_query, (collection_name, product_id))
        
       
    conn.commit()
    print("The 'collection' column is updated.")
except mysql.connector.Error as error:
        print("Failed to update records to nike db: ", error)

cursor.close()
conn.close()
print("MySQL connection is closed.")

```
